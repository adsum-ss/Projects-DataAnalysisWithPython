# WELCOME!

Welcome to **"Bike Demand Visualization Project"** which is the capstone project of Data Visualization Lessons . As you know recently, free or affordable access to bicycles has been provided for short-distance trips in an urban area as an alternative to motorized public transport or private vehicles. Thus, it is aimed to reduce traffic congestion, noise and air pollution.  

The aim of this project is to reveal the current patterns in the data by showing the historical data of London bike shares with visualization tools.

This will allow us to X-ray the data as part of the EDA process before setting up a machine learning model.

# Determines



#### Features

- timestamp - timestamp field for grouping the data
- cnt - the count of a new bike shares
- t1 - real temperature in C
- t2 - temperature in C “feels like”
- hum - humidity in percentage
- wind_speed - wind speed in km/h
- weather_code - category of the weather
- is_holiday - boolean field - 1 holiday / 0 non holiday
- is_weekend - boolean field - 1 if the day is weekend
- season - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.

**"weather_code" category description:**
* 1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity 
* 2 = scattered clouds / few clouds 
* 3 = Broken clouds 
* 4 = Cloudy 
* 7 = Rain/ light Rain shower/ Light rain 
* 10 = rain with thunderstorm 
* 26 = snowfall 
* 94 = Freezing Fog

### 5.Look at the data type of each variable, transform `timestamp` in type, and set it as index.

```python
df.dtypes
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.set_index("timestamp", inplace=True)
df.head()
```

### 6.Make feature engineering. Extract new columns (day of the week, day of the month, hour, month, season, year etc.)

you can use strftime('%Y-%m') for year_month

```python
# f=lambda x: x.strftime('%Y-%m') --- strftime works with array
# df['year_month']=  f(df.index)
```

```python
df['year_month'] = df.index.strftime('%Y-%m')
df['year'] = df.index.year
df['month'] = df.index.month
df['day_of_month'] = df.index.day
df['day_of_week'] = df.index.weekday  # df.index.dayofweek
df['hour'] = df.index.hour
```

### 7.Visualize the correlation with a heatmap

```python
# for all variables

plt.figure(figsize=(15,10))
sns.heatmap(data=df.corr(), annot=True, cmap='coolwarm');

plt.xticks(fontsize=11)
plt.yticks(fontsize=11);
```

```python
# for the target variable

plt.figure(figsize=(2,7))

sns.heatmap(data=df.corr().loc[:, ['cnt']].sort_values(by='cnt', ascending=False), 
            annot_kws={'fontsize':11}, vmin=-1, vmax=1, annot=True, cmap='BrBG')

plt.xticks(fontsize=11)
plt.yticks(fontsize=11);
```

### 15.Visualize the distribution of bike shares by weekday/weekend with piechart and barplot

```python
df['is_weekend'].value_counts()
```

```python
# pie chart

plt.figure(figsize=(5,5))

plt.pie(df['is_weekend'].value_counts(), labels=['weekday', 'weekend'], autopct='%.1f');
```

```python
# with countplot-1

plt.figure(figsize=(7,5))

x = df['is_weekend'].value_counts().index
y = df['is_weekend'].value_counts().values

sns.countplot(x='is_weekend', data=df)

plt.ylim(0, 14000)
plt.title('Distribution of Bike Share by Weekend', pad=10)

for i in range(len(x)):
        plt.text(x[i], y[i]+100, f'%{round((y[i]/y.sum())*100,1)} - {str(y[i])}', ha='center', va='bottom')
```

```python
# with countplot-2

fig, ax = plt.subplots(figsize=(7,5))
ax = sns.countplot(x='is_weekend',data=df)

ax.set_ylim(0, 14000)
ax.set_title('Distribution of Bike Share by Weekend', pad=10)

for bar in ax.patches: 
    ax.annotate('%' + str(round(100*bar.get_height()/len(df),1)) +' - '+ str(bar.get_height() ), 
                (bar.get_x()+0.25 , bar.get_height()+200), size=10)
```
